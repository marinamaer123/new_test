6.РЈ Р’Р°СҒ С” РІРөР»РёРәРёР№ DataFrame df, СүРҫ РҝСҖРөРҙСҒСӮР°РІР»СҸС” СӮСҖР°РҪР·Р°РәСҶС–С— РҝСҖРҫРҙР°Р¶Сғ РІ РјР°РіР°Р·РёРҪС–. РҶ В«product_nameВ», С– В«product_categoryВ» С” СҖСҸРҙРәР°РјРё, СүРҫ РҝСҖРөРҙСҒСӮР°РІР»СҸСҺСӮСҢ СҮРёСҒР»РөРҪРҪС–, Р°Р»Рө РҫРұРјРөР¶РөРҪС– СғРҪС–РәР°Р»СҢРҪС– РҪР°Р·РІРё РҝСҖРҫРҙСғРәСӮС–РІ С– РәР°СӮРөРіРҫСҖС–С— РІС–РҙРҝРҫРІС–РҙРҪРҫ.

РқР°РҝРёСҲС–СӮСҢ РәРҫРҙ РҙР»СҸ РөС„РөРәСӮРёРІРҪРҫРіРҫ РҝРөСҖРөСӮРІРҫСҖРөРҪРҪСҸ СҒСӮРҫРІРҝСҶС–РІ В«product_nameВ» С– В«product_categoryВ» РҪР° РәР°СӮРөРіРҫСҖС–Р№РҪРёР№ СӮРёРҝ РҙР»СҸ РөС„РөРәСӮРёРІРҪРҫРіРҫ РІРёРәРҫСҖРёСҒСӮР°РҪРҪСҸ РҝР°РјвҖҷСҸСӮС–.
РЎРҝСҖРҫРұСғР№СӮРө РІРёСҖС–СҲРёСӮРё СҶРө Р·Р°РІРҙР°РҪРҪСҸ, РјС–РҪС–РјС–Р·СғСҺСҮРё РІРёРәРҫСҖРёСҒСӮР°РҪРҪСҸ РҝР°Рј'СҸСӮС– СӮР° Р·РұРөСҖС–РіР°СҺСҮРё СҮРёСӮР°РұРөР»СҢРҪС–СҒСӮСҢ РәРҫРҙСғ. РҡСҖС–Рј СӮРҫРіРҫ, РІРёРјС–СҖСҸР№СӮРө РІРёРәРҫСҖРёСҒСӮР°РҪРҪСҸ РҝР°РјвҖҷСҸСӮС– РҙРҫ СӮР° РҝС–СҒР»СҸ РҝРөСҖРөСӮРІРҫСҖРөРҪРҪСҸ, СүРҫРұ РҝСҖРҫРҙРөРјРҫРҪСҒСӮСҖСғРІР°СӮРё РҝС–РҙРІРёСүРөРҪРҪСҸ РөС„РөРәСӮРёРІРҪРҫСҒСӮС–.

РҹРҫСҖС–РІРҪСҸР№СӮРө РәС–Р»СҢРәС–СҒСӮСҢ РҝР°РјКјСҸСӮС–, СҸРәСғ Р·Р°Р№РјР°С” РҪР°РҙР°РҪРёР№ РҪР°РұС–СҖ РҙР°РҪРёС… С– РҪРҫРІРёР№, СҒСӮРІРҫСҖРөРҪРёР№ Р’Р°РјРё. Р’ Р»РөРәСҶС–С— РұСғРІ РҝСҖРёРәР»Р°Рҙ. Р’ СҒРәС–Р»СҢРәРё СҖР°Р·С–РІ РјРөРҪСҲРө РҝР°РјКјСҸСӮС– Р·Р°Р№РјР°С” РҪРҫРІРёР№ РҙР°СӮР°С„СҖРөР№Рј?

````python
import pandas as pd
import numpy as np

np.random.seed(0)

products = ['apple', 'banana', 'cherry', 'dates', 'elderberry']
categories = ['fruit', 'fruit', 'fruit', 'fruit', 'fruit']

N_ROWS = 10**6
df = pd.DataFrame({
    'product_name': np.random.choice(products, N_ROWS),
    'product_category': np.random.choice(categories, N_ROWS),
    'product_price': np.random.uniform(1, 10, N_ROWS),
    'transaction_date': pd.date_range(start='1/1/2000', periods=N_ROWS, freq='h')
})
````

````python
before = df.memory_usage(deep=True).sum()

df['product_name'] = df['product_name'].astype('category')
df['product_category'] = df['product_category'].astype('category')

after = df.memory_usage(deep=True).sum()

print(f"Before: {before / 1e6:.2f} MB")
print(f"After: {after /1e6:.2f} MB")
print(f"Reduced by: {before /after:.2f}x")
````

````python
def CalculateVariance(nl):
    sum_list = 0
    for n in nl:
        sum_list = sum_list + n
    mean = sum_list / len(nl)
    sum_squares = 0
    for n in nl:
        sum_squares = sum_squares + n**2
    mean_squares = sum_squares / len(nl)
    return mean_squares
````
