```python
Area_dummies = pd.get_dummies(clean_df['Area Type'], prefix='')
Area_dummies.head().astype(int)
```

```python
City_dummies = pd.get_dummies(clean_df['City'], prefix='City')
City_dummies.head().astype(int)
```

```python
Furnishing_dummies = pd.get_dummies(clean_df['Furnishing Status'], prefix='')
Furnishing_dummies.head().astype(int)
```

```python
Tenant_dummies = pd.get_dummies(clean_df['Tenant Preferred'], prefix='Tenant')
Tenant_dummies.head().astype(int)
```

## Завдання 5: Розділення даних та навчання моделі (3 бали)

**Що потрібно зробити:**
1. Розділіть дані на навчальну (80%) та тестову (20%) вибірки.

5. Зробіть прогнози на тренувальній та тестовій вибірках.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled_df, Y,
    test_size=0.2,
    random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
for feature, weight in zip(model.feature_names_in_, model.coef_):
    print(f"{feature}: {weight:.2f}")

print(f"\nЗміщення (intercept): {model.intercept_:.2f}")
```

Найбільше впливають City_Mumbai та Size.

---



```python
y_train_pred = model.predict(X_train)


y_test_pred = model.predict(X_test)


comparison = pd.DataFrame({
    'Реальні Ціни': y_test.values[:10],
    'Прогнозовані Ціни': y_test_pred[:10].round(0),
    'Помилка': (y_test.values[:10] - y_test_pred[:10]).round(0)
})
print("Приклади прогнозів на тестовій вибірці:")
print(comparison)
```

## Завдання 6: Оцінка якості моделі (2 бали)

**Що потрібно зробити:**
1. Обчисліть MAE, RMSE та R² для навчальної та тестової вибірок
2. Порівняйте метрики та зробіть висновок про якість моделі
3. Проаналізуйте і дайте висновок, чи є ознаки перенавчання або недонавчання (**Нагадування**: перенавчання - коли модель дуже добре працює на тренувальних даних, але погано на тестових; недонавчання - коли модель погано працює навіть на тренувальних даних)
4. Побудуйте графік розсіювання "реальні vs прогнозовані значення" та зробіть висновок про якість моделі


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y_test,y_test_pred)
mse = mean_squared_error(y_test,y_test_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_test_pred)
print("="*50)
print("МЕТРИКИ ЯКОСТІ МОДЕЛІ (на тестовій вибірці):")
print("="*50)
print(f"\nMAE: {mae:.2f} ")
print(f"RMSE: {rmse:.2f} ")
print(f"R²: {r2:.3f}")
```

```python
mae2 = mean_absolute_error(y_train,y_train_pred)
mse2 = mean_squared_error(y_train,y_train_pred)
rmse2 = np.sqrt(mse)
r22 = r2_score(y_train,y_train_pred)
print("="*50)
print("МЕТРИКИ ЯКОСТІ МОДЕЛІ (на тестовій вибірці):")
print("="*50)
print(f"\nMAE: {mae2:.2f} ")
print(f"RMSE: {rmse2:.2f} ")
print(f"R²: {r22:.3f}")
```

При порівнянні R² ми бічимо що модель працює доволі добре,попри те різниця RMSE з MAE  вказує на те що модель в деяких випадках помиляється доволі сильно.Можливо на дуже дешевих та дуже дорогих приміщеннях. Ознак перенавчання або недонвчання не виявлено.

```python
fig = px.scatter(
    x=y_test,
    y=y_test_pred,
    title='Реальні vs Прогнозовані ціни (тестова вибірка)',
    labels={'x': 'Реальні ціни', 'y': 'Прогнозовані ціни'},
    opacity=0.6
)

# Додаємо ідеальну лінію (де прогноз = реальність)
max_val = max(y_test.max(), y_test_pred.max())
fig.add_trace(
    go.Scatter(
        x=[0, max_val],
        y=[0, max_val],
        mode='lines',
        name='Ідеальний прогноз',
        line=dict(color='red', dash='dash')
    )
)

fig.update_layout(height=500)
fig.show()
```
